## Loading Cdataset

```python
import pandas as pd
import scipy.io
import numpy as np
```

```python
# Load the MATLAB file
disease_data = scipy.io.loadmat('../../MBiRW/Datasets/CDatasets/DiseaseSimMat.mat')
disease_mat = disease_data['Wdd']
disease_names = []
with open('../../MBiRW/Datasets/CDatasets/DiseasesName', 'r') as f:
    disease_names = [line.strip() for line in f.readlines()]
```

```python
# Load the MATLAB file
drug_data = scipy.io.loadmat('../../MBiRW/Datasets/CDatasets/DrugSimMat.mat')
drug_mat = drug_data['test']
drug_names = []
with open('../../MBiRW/Datasets/CDatasets/DrugsName', 'r') as f:
    drug_names = [line.strip() for line in f.readlines()]
```

```python
disease_df = pd.DataFrame(disease_mat, index=disease_names, columns=disease_names)
drug_df = pd.DataFrame(drug_mat, index=drug_names, columns=drug_names)
```

```python
disease_df.to_csv("../data/Cdata_Lssrl/Cdataset_disease.csv")
drug_df.to_csv("../data/Cdata_Lssrl/Cdataset_drug.csv")
```

```python
drug_disease = pd.read_csv("../../MBiRW/Datasets/CDatasets/DrDiMat", sep="\t", header=None)
```

```python
drug_disease = drug_disease.drop(columns=409)
```

```python
drug_disease.index = drug_names
drug_disease.columns = disease_names
```

```python
drug_disease.to_csv("../data/Cdata_Lssrl/Cdataset_drug_dis.csv")
```

```python
drug_disease.info()
```

```python
(1 - (drug_disease.sum().sum())/(len(drug_names)*len(disease_names)))*100
```
