```python
interaction_set = set()
for i in interactions.values:
    drugi, diseasei = i[0], i[1]
    interaction_set.add((drugi,diseasei))
```

```python
from random import randint

non_interaction_set = set()

while len(non_interaction_set) != 10000:
    random_integer = randint(0, len(unique_drugs)-1)
    drug = (unique_drugs[random_integer])
    random_integer = randint(0, len(unique_diseases)-1)
    disease = (unique_diseases[random_integer])
    if (drug,disease) not in interaction_set:
        non_interaction_set.add((drug,disease))
```

```python
non_interaction_list = list(non_interaction_set)
interaction_list = list(interaction_set)
```

```python
input_matrix = []
for drug_disease in non_interaction_list:
    drug,disease = drug_disease
    drug_index = (unique_drugs.index(drug))
    disease_index = (unique_diseases.index(disease))
    result = np.concatenate([drug_snf[drug_index], disease_snf[disease_index], np.array([0])])
    input_matrix.append(result)
for drug_disease in interaction_list:
    drug,disease = drug_disease
    drug_index = (unique_drugs.index(drug))
    disease_index = (unique_diseases.index(disease))
    result = np.concatenate([drug_snf[drug_index], disease_snf[disease_index], np.array([1])])
    input_matrix.append(result)
X = np.array(input_matrix)
```
