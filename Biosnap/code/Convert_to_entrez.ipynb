## Code to read thorugh the knowledge graph to generate drug-gene dictionary and disease-gene dictionary 

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# loading the knowledge graph
network = pd.read_csv('../data/DG-Miner_miner-disease-gene.tsv', sep = "\t")
diseases = network['# Disease(MESH)'] 
# diseases = network['#Drug'] 
network.columns = ['a','b']
network
```

```python
# Loading the data of gene name mapping
reference = pd.read_csv('../data/reference.csv', index_col=0)
reference.index = reference['1']
reference = reference[~reference.index.duplicated(keep='first')]
```

```python
reference
```

```python
print(len(set(reference['1'])))
```

```python
val_list = []
for i, j in tqdm(network.iterrows()):
    if j['b'] in reference.index:
        val_list.append(reference.loc[j['b'],'0'])
    else:
        val_list.append("null")
```

```python
val_list
```

```python
network['c'] = val_list
```

```python
network.to_csv('../data/Disease_gene_net.csv')
```
