```python
import pandas as pd
```

```python
reference_path = "G-SynMiner_miner-geneHUGO.tsv"
network_path = "HumanNet-PI.tsv"
```

```python
reference = pd.read_csv(reference_path, sep = "\t")
```

```python
reference
```

```python
reference_list = []
for i, row in reference.iterrows():
    if str(row['entrez_id']).lower() != 'nan' and str(row['uniprot_ids']).lower() != 'nan':
        reference_list.append([int(row['entrez_id']),row['uniprot_ids'], row['symbol'] ])
```

```python
reference_filtered = pd.DataFrame(reference_list)
reference_filtered.to_csv("./reference.csv")
```
