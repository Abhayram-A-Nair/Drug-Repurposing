```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
drugg = pd.read_csv('Drug_gene_net.csv')
diseaseg = pd.read_csv('Disease_gene_net.csv')
```

```python
drugg
```

```python
diseaseg
```

```python
df = diseaseg.dropna(subset=['c'])
df
```

```python
gene_dict_disease = df.groupby('a')['c'].apply(list).to_dict()
gene_dict_disease
```

```python
df = drugg.dropna(subset=['c'])
df
```

```python
gene_dict_drug = df.groupby('a')['c'].apply(list).to_dict()
gene_dict_drug
```

```python
import json
with open('gene_dict_drug.json', 'w') as file:
    json.dump(gene_dict_drug, file)
with open('gene_dict_disease.json', 'w') as file:
    json.dump(gene_dict_disease, file)
```
