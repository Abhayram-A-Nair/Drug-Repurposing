```python
disease_data = pd.read_csv("../../data/SND/disease_snf14.csv", index_col=0 )
```

```python
### Run if the dataset is SND only
with open('../../data/SND/SNF_diseases_matched.json', 'r') as file:
    snf_disease = json.load(file)


# Loop through DataFrame's index, and set new index based on dictionary or random number
for index in disease_data.index:
    if index in snf_disease:
        disease_data = disease_data.rename(index={index: snf_disease[index]})
    else:
        disease_data = disease_data.rename(index={index: index})

disease_data
```

```python
disease_names = list(disease_data.index)
```

```python
# load the dictionary from the file
with open('../data/gene_dict_disease.json', 'r') as file:
    gene_dict_disease = json.load(file)
```

```python
count = 0 
for i in disease_names:
    if i in gene_dict_disease:
        count+=1
print(count)
```

```python
len(disease_names)
```

```python
disease_present = []
for disease in disease_names:
    if disease in gene_dict_disease and gene_dict_disease[disease] != []:
        disease_present.append(disease)
drug_present = []
for drug in drug_names:
    if drug in gene_dict_drug and gene_dict_drug[drug] != []:
        drug_present.append(drug)
```

```python
len(disease_present) , len(drug_present)
```

```python
new_drug_simmat = np.zeros((len(drug_present),len(drug_present)))
indices = np.zeros(len(drug_present), dtype = int)
cur = 0
for i, drug in enumerate(drug_names):
    if(drug in drug_present):
        indices[cur] = i
        cur+=1
        
cur = 0
for i in indices:
    new_drug_simmat[cur] = drug_simmat[i, indices]
    cur+=1

print(new_drug_simmat,new_drug_simmat.shape)
drug_indices = indices
```

```python
disease_simmat = disease_data.values
print(disease_simmat, disease_simmat.shape)
```

```python
new_disease_simmat = np.zeros((len(disease_present),len(disease_present)))
indices = np.zeros(len(disease_present), dtype = int)
cur = 0
for i, disease in enumerate(disease_names):
    if(disease in disease_present):
        indices[cur] = i
        cur+=1
        
cur = 0
for i in indices:
    new_disease_simmat[cur] = disease_simmat[i, indices]
    cur+=1

print(new_disease_simmat,new_disease_simmat.shape)
disease_indices = indices
```

```python
drug_df = pd.DataFrame(new_drug_simmat, index = drug_names[drug_indices], columns= drug_names[drug_indices])
drug_df
```

```python
disease_df = pd.DataFrame(new_disease_simmat, index = np.array(disease_names)[disease_indices], columns= np.array(disease_names)[disease_indices])
disease_df
```

```python
drug_df.to_csv('../../data/processed_drug_simmat_SNF19.csv')
disease_df.to_csv('../../data/processed_disease_simmat_SNF14.csv')
```
